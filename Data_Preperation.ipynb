```python
Raw_DIR = r'C:\Users\Avinash\Desktop\Major_Project\Drowsiness_detection\mrlEyes_2018_01\mrlEyes_2018_01'

for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split('_')[4] == '0':
            shutil.copy(src=dirpath+'/'+i, dst=r'C:\Users\Avinash\Desktop\Major_Project\Drowsiness_detection\mrlEyes_2018_01\Prepared_Data\Close_Eyes')
            
        elif i.split('_')[4] == '1':
            shutil.copy(src=dirpath+'/'+i, dst=r'C:\Users\Avinash\Desktop\Major_Project\Drowsiness_detection\mrlEyes_2018_01\Prepared_Data\Open_Eyes')
```

```python
#Transferred 4000 images of Open and close class to test dir 

#Open Eyes
source_folder = r'C:\Users\Avinash\Desktop\Major_Project\Drowsiness_detection\mrlEyes_2018_01\Prepared_Data\Open_Eyesss'
destination_folder = r'C:\Users\Avinash\Desktop\Major_Project\Drowsiness_detection\mrlEyes_2018_01\Prepared_Data\test\Open_Eyesss'
number_of_files_to_move = 4000


# Get a list of files in the source folder
files_to_move = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]

# Limit the number of files to move to the specified amount
files_to_move = files_to_move[:number_of_files_to_move]

for file_name in files_to_move:
    source_file_path = os.path.join(source_folder, file_name)
    destination_file_path = os.path.join(destination_folder, file_name)
    
    # Move the file
    shutil.move(source_file_path, destination_file_path)
    print(f"Moved: {source_file_path} -> {destination_file_path}")

print(f"Moved {len(files_to_move)} files.")
+

#Close Eyes
#Open Eyes
source_folder = r'C:\Users\Avinash\Desktop\Major_Project\Drowsiness_detection\mrlEyes_2018_01\Prepared_Data\Close_Eyesss'
destination_folder = r'C:\Users\Avinash\Desktop\Major_Project\Drowsiness_detection\mrlEyes_2018_01\Prepared_Data\test\Close_Eyesss'
number_of_files_to_move = 4000


# Get a list of files in the source folder
files_to_move = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]

# Limit the number of files to move to the specified amount
files_to_move = files_to_move[:number_of_files_to_move]

for file_name in files_to_move:
    source_file_path = os.path.join(source_folder, file_name)
    destination_file_path = os.path.join(destination_folder, file_name)
    
    # Move the file
    shutil.move(source_file_path, destination_file_path)

print(f"Moved {len(files_to_move)} files.")
```
